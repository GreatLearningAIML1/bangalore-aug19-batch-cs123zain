## Question 7

### Define weight and bias
- Initialize weight and bias with tf.zeros
- tf.zeros is an initializer that generates tensors initialized to 0
- Specify the value for shape

```python
W = tf.zeros(shape=(4, 1)) 
b = tf.zeros(shape=(1))
```

## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```python
@tf.function 
def prediction(features, W, b):     
  y_pred = tf.add(tf.matmul(features, W), b)     
  return y_pred
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```python
@tf.function 
def loss(y_actual, y_pred):     
  diff = y_actual - y_pred     
  sqre = tf.square(diff)     
  average = tf.reduce_mean(sqre)     
  return average
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```python
def train(x, y_act, w, b, lr=0.01):          
  # Record mathematical operations on 'tape' to calculate loss     
  with tf.GradientTape() as t:         
    t.watch([w,b])         
    current_pred = prediction(x, w, b)         
    current_loss = loss(y_act, current_pred)          
    
    # Calculate Gradients for Loss with respect to Weights and Bias     
    dw, db = t.gradient(current_loss,[w, b])          
    
    # Update Weights and Bias     
    w = w - lr * dw     
    b = b - lr * db          
    return w, b     
```

## Question 10

### Train the model for 100 epochs 
- Observe the training loss at every iteration

```python
for i in range(100):         
      W, b = train(x, y, W, b)     
      print('Current Training Loss on iteration is:', i, loss(y, prediction(x, W, b)).numpy())
```
