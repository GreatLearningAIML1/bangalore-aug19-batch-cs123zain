## 1.Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
laligaD=pd.read_csv("Laliga.csv")
laligaDF=pd.DataFrame(laligaD)
laligaDF=laligaDF.replace(regex=[r'\-[0-9]+',r'\-'], value=['','0'])
laligaDF=laligaDF.astype({'Debut': 'int32','Points': 'int32','GoalsAgainst':'int32','GoalsFor':'int32','GamesWon':'int32','GamesPlayed':'int32'},inplace=True)
laligaDF
#laligaDF.dtypes


```

## 4.Write a function with name â€œGoal_diff_countâ€ which should 
     4a. return all the teams with their Goal Differences. 
     4b. Using the same function, find the team which has maximum and minimum goal difference. 
<p>Goal_diff_count = GoalsFor - GoalsAgainst</p>

### 4a. Goal Differences

```python
def Goal_diff_count(min_max):
    if min_max==False:
        for row in laligaDF.iterrows():
            print('Team ',row[1]['Team'],"\n \t Goal For ",row[1]['GoalsFor']," \t Goals Against ",row[1]['GoalsAgainst'],' \t Difference ',row[1]['GoalsFor']-row[1]['GoalsAgainst'])   
    if min_max==True:
        laligaDF['GF_GA_Diff']=laligaDF['GoalsFor']-laligaDF['GoalsAgainst']
        print('minimum goal difference ',laligaDF['GF_GA_Diff'].min(),'\nmaximum goal difference ',laligaDF['GF_GA_Diff'].max())

Goal_diff_count(False)  


#laligaDF.dtype
```

###  4b.maximum and minimum goal difference

```python
Goal_diff_count(True)
```

### 5a.Percentage of Winning = (GamesWon / GamesPlayed)*100,If there are any numerical error,replace it with 0%

```python
laligaDF['Winning Percent']=(laligaDF['GamesWon']/laligaDF['GamesPlayed'])*100
laligaDF['Winning Percent'].fillna(0,inplace=True)
```

### 5b.Print the top 5 teams which has the highest Winning percentag

```python
laligaDF.sort_values(by=['Winning Percent'],axis=0,ascending=False).head(5)
```
